```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('data/NRI_Table_Counties.csv')
print("===============Data Info===============")
print(data.info())
```

```python
print("===============Summary Statistics===============")
print(data.describe())
```

```python

print("===============Data Head===============")
print(data.head())
print("===============Data Tail===============")
print(data.tail())
```

```python
print("===============Data Columns===============")
print(data.columns)
```

```python
print("===============Data Missing Values===============")
print(data.isnull().sum())
```

```python
print("========Unique values in categorical columns========")
categorical_columns = data.select_dtypes(include=['object']).columns
for column in categorical_columns:
    print(f"\n{column}:")
    print(data[column].value_counts())
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['RISK_SCORE'], kde=True)
plt.title("Distribution of Risk Scores")
plt.xlabel("Risk Score")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['BUILDVALUE'], kde=True)
plt.title("Distribution of Building Value")
plt.xlabel("Building Value")
plt.ylabel("Count")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['AGRIVALUE'], kde=True)
plt.title("Distribution of Agricultural Value")
plt.xlabel("Agricultural Value")
plt.ylabel("Count")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data['EAL_VALT'], kde=True)
plt.title("Distribution of Expected Annual Loss (Total)")
plt.xlabel("Expected Annual Loss (Total)")
plt.ylabel("Count")
plt.ticklabel_format(style='plain', axis='x')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='BUILDVALUE', y='EAL_VALT')
plt.title("Building Value vs. Expected Annual Loss (Total)")
plt.xlabel("Building Value")
plt.ylabel("Expected Annual Loss (Total)")
plt.ticklabel_format(style='plain', axis='both')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=data, x='AGRIVALUE', y='EAL_VALT')
plt.title("Agricultural Value vs. Expected Annual Loss (Total)")
plt.xlabel("Agricultural Value")
plt.ylabel("Expected Annual Loss (Total)")
plt.ticklabel_format(style='plain', axis='both')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=data[['BUILDVALUE', 'AGRIVALUE', 'EAL_VALT']])
plt.title("Comparison of Building Value, Agricultural Value, and Expected Annual Loss (Total)")
plt.ylabel("Value")
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)
plt.show()
```

```python
variables = ['POPULATION', 'BUILDVALUE', 'AGRIVALUE', 'AREA', 'RISK_VALUE', 'RISK_SCORE']

correlation_matrix = data[variables].corr()

print("Correlation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', square=True)
plt.title("Correlation Matrix")
plt.show()


print("\nPairwise Correlations:")
for i in range(len(variables)):
    for j in range(i+1, len(variables)):
        var1 = variables[i]
        var2 = variables[j]
        correlation = data[var1].corr(data[var2])
        print(f"Correlation between {var1} and {var2}: {correlation:.2f}")
```
